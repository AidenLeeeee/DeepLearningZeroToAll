# 1. Multinomial Classification

##### Logistic Classification


- 우리는 이전 파일에서 Logistic Regression 에 대해 학습했다
- Logistic Regression 은 Binary Classification, 즉 이진 분류 모델이었다
- 이는 데이터를 0 과 1 의 두 그룹으로 나눌 때 유용했다
- 하지만 Multinomial Classification 은 데이터를 셋 이상의 그룹으로 나누기 위한 모델이다
- 기본적으로 Logistic Classification 을 확장시켜 이해하면 쉽게 이해할 수 있다



- Logistic Regression 의 과정은 위와 같다
- X 가 주어지고 Hypothesis 를 도출 
- 해당 값을 Sigmoid Function 을 통해 0 ~ 1 사이의 값으로 반환
- 이 때, Y 는 Prediction (예측값) 으로서, Y-hat 으로 표현한다

##### Multinomial Classification


- 위의 데이터를 설명하자면, 이렇다
- x1 은 공부한 시간(hours)
- x2 는 수업 참석횟수(attendance)
- y 는 성적(grade)



- 위의 데이터를 그래프로 표현하였다
- 데이터는 총 세 개의 클래스(A, B, C) 로 나누어져 있으며
- 각 y 값에 따라 데이터들이 군집을 이루었다
- 이러한 데이터는 Binary Classification 을 사용해 충분히 구현할 수 있을 것이다
- (A or notA), (B or notB), (C or notC) 의 각 Binary Classification 을 사용하면 된다

- 이해를 돕기 위해 3 개의 Binary Classification 그래프를 가져왔다
- A 인지 아닌지를 판단하는 Binary Classification (A or notA)
- B 인지 아닌지를 판단하는 Binary Classification (B or notB)
- C 인지 아닌지를 판단하는 Binary Classification (C or notC)
- 이렇게 세 개의 Binary Classification 으로 A, B, C 세 개의 클래스로 분류할 수 있는 것이다


- 이를 위해 우리는 각각 독립된 세 경우(A or notA), (B or notB), (C or notC) 의 Y-hat 을 구해야 한다


- 그리고 독립된 Y-hat 을 구하는 세 경우의 식은 위와 같을 것이다


- 또한, 독립된 세 개의 수식을 위와 같이 하나의 Vector 로 묶어 처리할 수 있다
- 이렇게 표현하는 방법이 곧 Multinomial Classification 이다
- Logistic Regression 을 토대로 다중 클래스가 있는 경우 각각의 수식을 묶어 하나의 Vector 로 표현하고
- 이를 통해, 한 번에 계산이 가능하고 세 개의 독립된 Classification 처럼 동작을 한다
