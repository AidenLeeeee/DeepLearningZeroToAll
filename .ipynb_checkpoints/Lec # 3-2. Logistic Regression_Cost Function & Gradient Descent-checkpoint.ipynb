# 1. Cost Function

##### Linear Regression

- Linear Regression 에서 Hypothesis 는 위와 같은 식으로 표현할 수 있었으며,

- 이에 따라, Cost Function 을 위와 같은 식으로 도출하였다

##### Logistic Regression

- 하지만 Logistic Regression 에서 Hypothesis 는 위와 같은 식으로 사용하게 된다
- 따라서, Logistic Regression 에서의 Cost Function 수식 또한 바뀌게 될 것이다
- 그렇다면, Linear Regression 의 Cost Function 에 위의 Hypothesis 수식을 대입하기만 하면 될까?

##### Error
- Linear Regression 의 Cost Function 에 Logistic Regression 의 Hypothesis 를 대입한 경우

- 위와 같은 비용함수의 그래프가 도출된다
- 위의 그래프에서 Gradient Descent 를 사용하게 되면 Global Minimum 이 아닌 Local Minimum 이 결과로 도출될 위험이 있다
- 이러한 문제점 때문에 Logistic Regression 에서는 새로운 Cost Function 수식이 필요하다

##### Solution

- 위의 수식은 Logistic Regression 의 Cost Function 수식이다
- 특정 수식으로 도출된 값들을 모두 합하고 데이터의 수로 나누는 logic 까지는 Linear 와 동일하다
- 하지만, C 함수가 새롭게 추가되었으며,
- y = 1, y = 0 의 두 가지 경우에 각각 C 함수를 다르게 적용하여 값을 도출한다
- 자연상수 e 와 상극인 log 를 사용하여 그래프의 굴곡을 없애고 Convex 한 그래프를 만들기 위함이다

##### y = 1

- y = 1 일 경우에 C(H(x), y) 의 그래프이다
- Hypothesis 가 1 일 경우, Cost 는 0 이며
- Hypothesis 가 0 일 경우, Cost 는 무한대로 늘어나게 된다

##### y = 0

- y = 0 일 경우에 C(H(x), y) 의 그래프이다
- Hypothesis 가 1 일 경우, Cost 는 무한대로 늘어나게 되며
- Hypothesis 가 0 일 경우, Cost 는 0 이다

##### Conclusion


- 우리는 위와 같은 Logistic Regression 의 Cost Function 수식을 살펴보았다
- 위 수식은 0 ~ 1 의 Hypothesis 를 인수로 받아 그에 따른 Cost 를 돌려줄 수 있었으며,
- Cost 는 Hypothesis 와 실제 y 값 간 차이를 0 ~ 무한대의 값으로 반환해주었다
- log 를 활용하여 자연상수 e 로 인한 그래프의 굴곡을 상쇄시켜 Convex 그래프를 만들어낼 수도 있었다
- Convex 그래프에서는 어떤 시작점에서도 Global Minimum 을 찾아낼 수 있기 때문이다

- 하지만 항상 if 문을 달고 다니며 y 값이 1 인지 0 인지 판단하고 각각 다른 식을 적용하기는 어렵다
- 따라서, 우리는 위의 두 식을 하나의 식으로 아래와 같이 표현한다

- 만약 y = 1 을 대입하면, 뒤 식은 0 이 되고 -log(H(x)) 와 같은 식이 될 것이고
- 만약 y = 0 을 대입하면, 앞의 식은 0 이 되고 -log(1-H(x)) 와 같은 식이 될 것이다

- 결론적으로 우리는 아래와 같은 Logistic Regression 의 Cost Function 수식을 도출하였다


# 2. Gradient Descent Algorithm

- 앞서 살펴본 Cost Function 식을 통해 우리는 Convex 그래프를 완성했고 
- Gradient Descent Algorithm 을 사용하여 Cost 를 Minimize 할 수 있게 되었다
- Logistic Regression 의 Gradient Descent 는 Linear Regression 에서의 공식과 동일하다
- 단지 Gradient Descent 에서 사용되는 Cost Function 이 변경되었다는 점만 유의하면 될 것이다
