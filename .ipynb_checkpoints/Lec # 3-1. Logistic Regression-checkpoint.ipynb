# 2. Linear Regression VS Logistic Regression

##### 2-1) Linear Regression

- 위의 그래프는 이전에 학습한 Linear Regression 을 나타낸 그래프이다
- 데이터들이 연속적이며, 선형적으로 배치되어있어 측정할 수 있다는 특징을 가지고 있다
- 데이터를 이은 선은 각 실제 데이터를 가장 잘 나타낼 수 있는 Hypothesis 이다

##### 2-2) Logistic Regression

- 위의 그래프는 앞으로 학습할 Logistic Regression 을 나타낸 그래프이다
- 데이터들이 ■ 혹은 ▲, 두 클래스로 나뉘어져 있는 모습을 확인할 수 있다
- 데이터 사이의 선은 실제 데이터를 가장 잘 분류할 수 있는 기준의 Hypothesis 이다

##### 2-3) Error 1


- 위의 그래프는 공부 시간에 따른 Pass 와 Fail 을 분류하는 그래프이다
- 빨간색으로 표시된 점은 Fail 을, 파란색으로 표시된 점은 Pass 를 나타낸다
- 연두색으로 표시된 선은 데이터들에 대한 Hypothesis 이다
- 우리는 이 때, 선분 a 즉 4.5 hours 를 기준으로 그 이상은 Pass, 이하는 Fail 이라는 결과를 유도할 수 있다
- 이는 언뜻 보면 Linear Regression 모델을 Logistic Regression 에도 활용할 수 있을 것으로 보인다
- 하지만 문제는 가령, 큰 수치 값을 가진 특이 데이터가 입력되었을 경우이다


- 위는 50 hours 의 Pass 데이터가 입력되었을 경우의 그래프이다
- 수치가 큰 데이터에 의해 그래프의 기울기는 크게 변화하게 되고,
- Fail 과 Pass 를 분류하는 선분 a 는 선분 b 로 이동하게 된다
- 이에 따라, 특이 데이터에 의해 기존 5 hours 는 Fail 로 분류되게 되는 문제가 발생하게 되는 것이다

##### 2-4) Error 2


- 다음 발생하는 문제점은 위의 수식이다
- 이는 Linear Regression 에서 사용한 함수이다
- 하지만, Logistic Regression 에서 Y 값은 "0" 또는 "1", 2 개의 클래스로만 도출되어야 한다
- 위의 수식에 큰 수치의 데이터를 입력하게 되면, "0" 보다 훨씬 작거나, "1" 보다 훨씬 큰 값이 도출된다
- 따라서, "0" 과 "1" 의 클래스로 분류하는 모델에서 사용하기에는 적합하지 않은 수식이다

##### 2-5) Solution

##### ▶ Sigmoid Function

- 우리는 Sigmoid Function 을 통해 이러한 문제를 해결할 수 있다
- 위의 수식에서 인자 z 는 H(x), 즉 Wx + b 를 의미한다
- 즉, H(x)를 인수로 받는 g() 함수는 그 값을 0 에서 1 사이로 돌려주는 역할을 수행하는 것이다
- 이 때, g() 함수를 Sigmoid Function 이라 한다


- 위 수식을 Logistic Regression 의 Hypothesis 에 적용하면 위와 같은 식이 도출된다
- Linear Regression 의 Hypothesis 인 Wx + b 를 인자로 받아 0 ~ 1 의 출력값을 가지게 된다

##### ▶ Sigmoid Function Graph

- 위 그림은 Sigmoid Function 의 Graph 이다
- x축 상의 임의의 입력값에 대해 0 에 수렴하거나, 1 에 수렴하는 모습을 확인할 수 있다
- 결국, Sigmoid Function 의 출력값을 결과가 분류될 "확률"로서 바라볼 수 있다
- 0.5 보다 높은 값은 "1" 로 분류될 확률이 더 높고,
- 0.5 보다 낮은 값은 "0" 으로 분류될 확률이 높다고 볼 수 있는 것이다
