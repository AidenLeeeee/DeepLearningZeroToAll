# Linear Regression

## 1. Data


- 특정 데이터로 학습을 시킨다 (training data)
- 학습 데이터를 기반으로 x 값을 입력하면 y 값을 도출

## 2. Graph (Hypothesis)


- 해당 데이터를 그래프로 표현
- 선형 구조의 그래프로 표현되어 가설을 세워볼 수 있다
- 현실세계의 많은 상황들은 선형 구조로 표현될 수 있다


- 그래프의 학습 데이터를 가장 잘 나타낼 수 있는 가설을 찾아야 한다
- Linear Regression 의 첫 번째 단계로 일차방정식의 가설을 설정

## 3. Cost Function


- 가설과 실제 데이터 간 오차를 측정 및 비교
- 이를 "Cost function" "Loss function" 이라 한다
- (H(x) - y)^2 활용
- 오차를 항상 양수로 표현하며, 오차가 클 때 더 큰 패널티를 얻을 수 있다


##### Cost Function
1. 각 데이터의 가설과 실제값 간 오차를 제곱
2. 제곱한 값을 모두 더함
3. 데이터의 개수로 나눔

##### Cost Function 의 학습과 목적


- 이러한 Cost를 최소로 만드는 W와 b를 찾아내는 것이 학습이자 목표

 ##### Cost Function 의 원리


- W와 b를 변화시키며 Cost를 도출


- x축을 W, y축을 cost 로 하여 이차함수 그래프를 도출
- 이 때, cost가 가장 낮은 지점의 W를 기계적으로 찾아내는 것이 목표
- 해당 지점을 찾아내기 위해 Gradient Descent Algorithm 을 주로 사용

## 4. Gradient Descent Algorithm


- Cost Function 그래프에서 Cost를 minimize하는 대표적인 방법
- 그래프와 현재 좌표가 접점하는 "경사"가 0 이 되는 지점을 찾아야 한다

##### 현재 "경사" 를 구하는 방법 - Derivative


- Derivative 를 활용하여 그래프 상의 점에 대한 기울기를 도출할 수 있다


- 이를 식으로 나타내면 위의 식으로 나타낼 수 있다
- 위 식을 반복 실행하면 W 값이 변화되고, Cost 가 최소화되는 W 값을 도출할 수 있다

##### 주의사항


- W와 b, Cost를 나타낸 3차원 그래프이다
- 해당 그래프에서 Gradient Descent Algorithm 을 통해 Cost 를 최소화할 때 문제가 발생할 수 있다
- 찾아낸 W와 b의 값이 그래프 전체에서 Cost의 최저점, 
즉 Global Cost minimum 인지 확신할 수 없다
- 우리가 임의로 설정한 시작값에 따라 Global Cost minimum 에 도달할 수 없을 수도 있다는 의미이다


- 위의 그래프는 Convex Function 이다
- 어떤 점에서 출발하더라도 Global Cost minimum 을 도출해낼 수 있다
- 따라서, 항상 Cost Function 이 Convex Function 의 형태인지 확인하여야 한다
